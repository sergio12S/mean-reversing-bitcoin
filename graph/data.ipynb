```python
# create simple method with Python typing
from typing import List
def print_polarity(data: List, value: float, value_subj: float) -> List:
    messages = list()
    for i in range(len(data)):
        polarity = data[i].get('polarity')
        subjectivity = data[i].get('subjectivity')

        # 1 rules 
        if float(polarity[0]) >= value:
            # 2 rules
            if float(subjectivity[0]) >= value_subj:
                messages.append(data[i].get('text'))
    return messages


print_polarity(data=data, value=0.8, value_subj=0.5)
```
