```python
from backtester import Backtest
import pandas as pd
import numpy as np
import datetime
from strategies.dataset import Dataset
```

```python
data_binance = Dataset()
data = data_binance.get_data(days=180, ticker='BTCUSDT', ts='1d')
```

```python
# correction = dict({'start': datetime.datetime(2021, 3, 14), 'end': datetime.datetime(2021, 3, 25)})
# growth = dict({'start': datetime.datetime(2021, 3, 26), 'end': datetime.datetime(2021, 4, 15)})
def correction_fn(correction, growth, data):
    data.loc[:, 'abs'] = data['high'] - data['low']
    data_correction = data[correction.get('start'):correction.get('end')]['abs'].cumsum()[-1]
    data_growth = data[growth.get('start'):growth.get('end')]['abs'].cumsum()[-1]
    temp = dict({
        'chg': data_correction/data_growth,
        'distantion_correction': data_correction,
        'distantion_growth': data_growth
    })
    return temp
```

```python
list_correction = [
    '2021, 01, 10 - 2021, 01, 31',
    '2021, 02, 22 - 2021, 02, 28',
    '2021, 03, 14 - 2021, 03, 25'
]
growth_correction = [
    '2021, 02, 01 - 2021, 02, 21',
    '2021, 03, 01 - 2021, 03, 13',
    '2021, 03, 26 - 2021, 04, 15'
]
```

```python
distantion_coint = []
for l, g in zip(list_correction, growth_correction):
    l_correction_start = l.split('-')[0].split(',')
    l_correction_end = l.split('-')[1].split(',')
    g_growth = g.split('-')
    l_growth_start = g.split('-')[0].split(',')
    l_growth_end = g.split('-')[1].split(',')
    date_correction = dict(
        {
            'start': datetime.datetime(int(l_correction_start[0]), int(l_correction_start[1]), int(l_correction_start[2])),
            'end': datetime.datetime(int(l_correction_end[0]), int(l_correction_end[1]), int(l_correction_end[2]))
        })
    date_growth = dict(
        {
            'start': datetime.datetime(int(l_growth_start[0]), int(l_growth_start[1]), int(l_growth_start[2])),
            'end': datetime.datetime(int(l_growth_end[0]), int(l_growth_end[1]), int(l_growth_end[2]))
        })
    distantion_coint.append(correction_fn(correction=date_correction,growth=date_growth,data=data))
print(distantion_coint)   
    
```
