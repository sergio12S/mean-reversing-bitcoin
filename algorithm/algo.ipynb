```python
data = [1, 2, 4, 9, 12, 13, 15, 19, 20, 22, 99, 100, 200, 300, 400, 500]
search = 100
```

```python
def binary_search(data, item):
    low = 0
    high = len(data) - 1
    while low <= high:
        mid = high + low
        guess = data[mid]
        if guess == item:
            return mid
        elif guess > item:
            high = mid - 1
        else:
            low = mid + 1
    return None
```

```python
binary_search(data, 100)
```
