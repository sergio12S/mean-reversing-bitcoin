# The biggest transaction 

```python
freq_transactions = 50
data_freq = requests.get(f'https://crypto2data.com/api/whale/frequency/{50}/')
data_freq = json.loads(data_freq.text).get('data')
```

```python
# Find the biggest transaction
data_count = [i.get('count') for i in data_freq]
biggest_transaction = max(data_count)
print(biggest_transaction)
```

```python
data_biggest_transaction = [i for i in data_freq if i.get('count') == biggest_transaction]
```

```python
transactions = data_biggest_transaction[0].get('transactions')
all_blockchain = list({i.get('blockchain') for i in transactions})
all_volume = sum(i.get('amount_usd') for i in transactions)
all_transaction_type = list({i.get('transaction_type') for i in transactions})
timestamp = datetime.fromtimestamp(
    list({i.get('timestamp') for i in transactions})[0]
)

all_from_wallet = [i.get('from') for i in transactions]
all_from_wallet_owner = [i.get('owner') for i in all_from_wallet]

all_to_wallet = [i.get('to') for i in transactions]
all_to_wallet_owner = [i.get('to') for i in all_to_wallet]

# print(timestamp)
# print(all_volume)
# print(all_blockchain)
print(transactions[0].keys())
# print(transactions[0])
```
