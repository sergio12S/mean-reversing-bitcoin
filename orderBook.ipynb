```python
from binance.client import Client
client = Client('key', 'secret_key')
#https://algotrading101.com/learn/binance-python-api-guide/
```

```python
status = client.get_system_status()
```

```python
# Open Order
buy_order_limit = client.create_test_order(
    symbol='SFPUSDT',
    side='BUY',
    type='LIMIT',
    timeInForce='GTC',
    quantity=10,
    price=2.5)
print(buy_order_limit)
```

```python
# Get open orders
check_orders = client.get_open_orders(symbol='SFPUSDT')
```
