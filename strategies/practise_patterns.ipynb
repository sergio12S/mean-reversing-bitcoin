```python
# Create your owm rules
data_list = []
for index, values in data.tail(100).iterrows():
    data.loc[index, 'signal'] = 0
    rules_1 = (values.get('high') - values.get('low')) / values.get('high')
    if rules_1 > 0.005:
        data.loc[index, 'signal'] = 1
    data_list.append(rules_1)

data_np = np.array(data_list) 
```

```python
print(np.max(data_np))
print(np.min(data_np))
print(np.std(data_np))
```
